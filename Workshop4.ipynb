```python
import matplotlib.pyplot as plt


years = ['2017', '2018', '2019', '2020', '2021', '2022']

data = [
    [8, 150, 80],
    [54, 77, 54],
    [93, 32, 100],
    [116, 11, 76],
    [137, 6, 93],
    [184, 1, 72]
]

labels = ['Bears', 'Dolphins', 'Whales']


fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes = axes.flatten()


for i in range(6):
    axes[i].pie(data[i], labels=labels, autopct='%1.1f%%')
    axes[i].set_title(years[i])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Red','Silver','Blue','White','Green']
y = [123,230,78,193,12]

colours = ["Magenta", "Green", "Cyan", "Red", "Yellow"]

plt.pie(
    y,
    labels=x,
    explode=[0.3, 0, 0, 0, 0],
    autopct='%1.2f%%',
    shadow=True,
    startangle=90,
    colors=colours
)

plt.legend()
plt.title('Pie Chart')

plt.show()
```
