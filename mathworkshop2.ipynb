```python
def unique_list(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique


sample = [1,2,3,3,3,3,4,5]
print(unique_list(sample))
```

```python
def prime(num):
    if num <= 1:
        return "Not prime"

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return "Not prime"

    return "Prime"

a = int(input("Enter a number: "))
print(prime(a))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True


for i in range(2, 1000):
    if is_prime(i) and is_prime(i + 2):
        print((i, i+2))
```

```python
def sumOfDivisor(num):
    total = 0

    for i in range(1, num):
        if num % i == 0:
            total += i

    if total == num:
        return "Perfect number"
    else:
        return "Not perfect number"

a = int(input("Enter a number: "))
print(sumOfDivisor(a))
```

```python
num = 7
def isPerfect(num):
    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i
    return total == num
while True:
    if isPerfect(num):
        print("Next perfect number is:", num)
        break
    num += 1
```

```python
marks = [20, 30, 35, 55, 70, 80, 90]


total = 0
for i in marks:
    total += i
mean = total / len(marks)

n = len(marks)


if n % 2 == 1:
    median = marks[n // 2]
else:
    mid1 = marks[n // 2 - 1]
    mid2 = marks[n // 2]
    median = (mid1 + mid2) / 2

print("Mean =", mean)
print("Median =", median)
```

```python
def decimal_to_binary(num):
    if num == 0:
        return "0"

    binary = ""

    while num > 0:
        remainder = num % 2
        binary = str(remainder) + binary
        num = num // 2

    return binary


n = int(input("Enter a decimal number: "))
print("Binary:", decimal_to_binary(n))
```

```python
def prodDigits(num):
    product = 1

    if num < 0:
        num = abs(num)

    if num == 0:
        return 0

    while num > 0:
        digit = num % 10
        product *= digit
        num //= 10
    return product


number = 12345
print(f"The product of digits of {number} is: {prodDigits(number)}")
```

```python
def multiply_args(*args):
    product = 1
    for num in args:
        product *= num
    return product


result_set1 = multiply_args(1, 2, 3, 4)
print(f"Product of (1, 2, 3, 4): {result_set1}")


result_set2 = multiply_args(5, 6, 7)
print(f"Product of (5, 6, 7): {result_set2}")


result_set3 = multiply_args(10, 20, 0.5, 2)
print(f"Product of (10, 20, 0.5, 2): {result_set3}")
```
