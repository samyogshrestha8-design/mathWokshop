```
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-10, 10, 20)

y1 = 5*x**3 + 2*x - 1
y2 = -2*x**3 + x**2 + 100
y3 = 2*math.pi*x + 20

plt.figure()

# Subplot 1
plt.subplot(3, 1, 1)
plt.plot(x, y1)
plt.title("5x^3 + 2x - 1")

# Subplot 2
plt.subplot(3, 1, 2)
plt.plot(x, y2)
plt.title("-2x^3 + x^2 + 100")

# Subplot 3
plt.subplot(3, 1, 3)
plt.plot(x, y3)
plt.title("2πx + 20")

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import math

a = 2
b = 2
r = 4

x_vals = []
y_vals = []

theta = 0


while theta <= 2 * math.pi:
    x = a + r * math.cos(theta)
    y = b + r * math.sin(theta)

    x_vals.append(x)
    y_vals.append(y)

    theta += 0.1


plt.plot(x_vals, y_vals)
plt.title("Circle with center (2,2) and radius 4")
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect('equal')
plt.grid()

plt.show()
```

```
import matplotlib.pyplot as plt
import math


a = 2
b = 2
r = 4


A = 6
B = 3

circle_x = []
circle_y = []

ellipse_x = []
ellipse_y = []

theta = 0


while theta <= 2 * math.pi:

    cx = a + r * math.cos(theta)
    cy = b + r * math.sin(theta)
    circle_x.append(cx)
    circle_y.append(cy)


    ex = a + A * math.cos(theta)
    ey = b + B * math.sin(theta)
    ellipse_x.append(ex)
    ellipse_y.append(ey)

    theta += 0.1


plt.plot(circle_x, circle_y, label="Circle")
plt.plot(ellipse_x, ellipse_y, label="Ellipse")

plt.title("Circle and Ellipse on Same Axes")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.gca().set_aspect('equal')

plt.show()
```
