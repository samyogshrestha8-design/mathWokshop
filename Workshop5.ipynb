```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create X values
x = np.linspace(-10, 10, 100)

# Step 2: Compute Y values
y = 4 * x**2 + 2

# Step 3: Plot the graph
plt.plot(x, y)
plt.title("Graph of f(x) = 4x^2 + 2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# origin
origin = [0, 0, 0]

# plot vectors
ax.quiver(*origin, *U)
ax.quiver(*origin, *V)
ax.quiver(*origin, *W)

ax.set_xlim([0, 1.5])
ax.set_ylim([0, 1.5])
ax.set_zlim([0, 1.5])

ax.set_title("3D Vectors")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

U = np.array([-0.5, 0.5])
V = np.array([1, 1])

W = U + V

origin = [0, 0]

plt.quiver(*origin, *U, angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, *V, angles='xy', scale_units='xy', scale=1)
plt.quiver(*origin, *W, angles='xy', scale_units='xy', scale=1)

plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.grid()
plt.title("2D Vector Addition")

plt.show()
```

```python
import numpy as np

def magnitude(v):
    return np.sqrt(np.sum(v**2))

U = np.array([1.5, 1.5, 2.0])
print(magnitude(U))
```

```python
def direction_cosines(v):
    mag = np.sqrt(np.sum(v**2))
    return v / mag

O = np.array([1.5, 1.5, 2.0])
print(direction_cosines(O))
```

```python
def dot_product(a, b):
    return np.dot(a, b)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])

print(dot_product(a, b))
```

```python
def cross_product(a, b):
    return np.cross(a, b)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])

print(cross_product(a, b))
```
